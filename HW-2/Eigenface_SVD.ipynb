```python
# Written Ex.3
import numpy as np
M = np.array([[1,0,3],[3,7,2],[2,-2,8],[0,-1,1],[5,8,7]])
A = np.dot(M.T,M)
B = np.dot(M,M.T)
print("(a):")
print("M.T*M=")
print(A)
print("M*M.T=")
print(B)

print("\n(b):")
print("Eigenvalues of M.T*M=")
print(np.linalg.eig(A)[0])
print("Eigenvalues of M*M.T=")
print(np.linalg.eig(B)[0])

print("\n(c):")
print("Eigenvectors of M.T*M=")
print(np.linalg.eig(A)[1])
print("Eigenvectors of M*M.T=")
print(np.linalg.eig(B)[1])

print("\n(d):")
print("SVD of M:")
U,E,Vh = np.linalg.svd(M)
print("U = eigenvectors of M*M.T = ")
print(U[:,:2])
print("Sigma = diag(sqrt(eigenvalues of M*M.T)) = ")
print(np.diag(E[:2]))
print("V.T = eigenvectors of M.T*M = ")
print(Vh[:2,:])
R = np.dot(np.dot(U[:,:2],np.diag(E[:2])),Vh[:2,:])
print("U * Sigma * V.T = ")
print(R)

print("\n(e):")
print("One-dimentinal approximation of M = ")
print(U[:,:1])
print("*")
print(np.diag(E[:1]))
print("*")
print(Vh[:1,:])
print("=")
R = np.dot(np.dot(U[:,:1],np.diag(E[:1])),Vh[:1,:])
print(R)
```

```python
import numpy as np
from scipy import misc
from matplotlib import pylab as plt
import matplotlib.cm as cm
%matplotlib inline

## load data
train_labels, train_data = [], []
for line in open('./faces/train.txt'):
    im = misc.imread(line.strip().split()[0])
    train_data.append(im.reshape(2500,))
    train_labels.append(line.strip().split()[1])
train_data, train_labels = np.array(train_data, dtype=float), np.array(train_labels, dtype=int)

test_labels, test_data = [], []
for line in open('./faces/test.txt'):
    im = misc.imread(line.strip().split()[0])
    test_data.append(im.reshape(2500,))
    test_labels.append(line.strip().split()[1])
test_data, test_labels = np.array(test_data, dtype=float), np.array(test_labels, dtype=int)

## Q(b)
print(train_data.shape, train_labels.shape)
print(test_data.shape, test_labels.shape)
plt.subplot(1,2,1)
plt.imshow(train_data[10, :].reshape(50,50), cmap = cm.Greys_r)
plt.axis('off')
plt.subplot(1,2,2)
plt.imshow(test_data[10, :].reshape(50,50), cmap = cm.Greys_r)
plt.axis('off')
plt.show()
```

```python
## Q(c)
train_aver_face = np.mean(train_data, 0)
plt.subplot(1,2,1)
plt.imshow(train_aver_face.reshape(50,50), cmap = cm.Greys_r)
plt.axis('off')
```

```python
## Q(d)
X_train = train_data - train_aver_face
X_test = test_data - train_aver_face
plt.subplot(1,2,1)
plt.imshow(X_train[10,:].reshape(50,50), cmap = cm.Greys_r)
plt.axis('off')
plt.subplot(1,2,2)
plt.imshow(X_test[10,:].reshape(50,50), cmap = cm.Greys_r)
plt.axis('off')
plt.show()
```

```python
## Q(e)
U, E, Vh = np.linalg.svd(X_train)
for i in range(10):
    plt.subplot(3,4,i+1)
    plt.imshow(Vh[i,:].reshape(50,50), cmap = cm.Greys_r)
    plt.axis('off')
plt.show()
```

```python
## Q(f)
err = np.zeros(200)
for r in np.arange(1, 201):
    err[r-1] = np.linalg.norm(X_train - np.dot(np.dot(U[:,:r], np.diag(E[:r])),Vh[:r,:]))
plt.plot(np.arange(1,201), err)
plt.show()
```

```python
## Q(g),(h)
# img_z = V*f => f = Vh*img_z (f:r-dim is the new img in axis Vh(r))
from sklearn.linear_model import LogisticRegression
r = 10
F = np.dot(X_train, Vh[:r,:].T) # 540*2500 * 2500*r
F_test = np.dot(X_test, Vh[:r,:].T) # 100*2500 * 2500*r
clf = LogisticRegression(solver='liblinear').fit(F, train_labels)
print(clf.predict(F_test))
print(clf.score(F_test, test_labels))
```

```python
## Q(h)
acc = np.zeros(200)
for r in np.arange(1,201):
    F = np.dot(X_train, Vh[:r,:].T) # 540*2500 * 2500*r
    F_test = np.dot(X_test, Vh[:r,:].T) # 100*2500 * 2500*r
    clf = LogisticRegression(solver='liblinear').fit(F, train_labels)
    acc[r-1] = clf.score(F_test, test_labels)
plt.plot(np.arange(1,201), acc)
plt.show()
```
